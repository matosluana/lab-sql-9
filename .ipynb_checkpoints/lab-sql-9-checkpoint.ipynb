```python
connection_string = 'mysql+pymysql://root:'+password+'@localhost/sakila'
engine = create_engine(connection_string)
data = pd.read_sql_query('SELECT * FROM rental', engine) # testing if connection works
data.head()
```

```python
# 1 - How many distinct (different) actors' last names are there?

query1 = 'SELECT COUNT(DISTINCT(last_name)) FROM sakila.actor'
data1 = pd.read_sql_query(query1, engine)
data1.head()
```

```python
# 2 - Add an additional column day_type with values 'weekend' and 'workday' depending on the rental day of the week.

query2 = '''SELECT *,
CASE
WHEN WEEKDAY (left(rental_date,12))<=4 then 'Weekday'
WHEN WEEKDAY(left(rental_date,12))>4 then 'Weekend'
ELSE 'No date information'
END AS 'rental_day_week'
FROM sakila.rental;'''
data2 = pd.read_sql_query(query2, engine)
data2.head()
```

```python
# 3 - Get all films with ARMAGEDDON in the title.
query3 = '''SELECT *
FROM sakila.film
WHERE title LIKE '%%armageddon%%';'''
# pd.set_option('display.max_rows', None)
data3 = pd.read_sql_query(query3, engine)
data3
```

```python
# 4 - Get 10 the longest films.

query4 = '''SELECT title, length 
FROM sakila.film
ORDER BY length DESC
LIMIT 10'''
data4 = pd.read_sql_query(query4, engine)
data4
```

```python
# 5 - How many films include Behind the Scenes content?

query5 = '''SELECT COUNT(title)
FROM sakila.film
WHERE special_features LIKE '%%Behind the Scenes%%';'''
data5 = pd.read_sql_query(query5, engine)
data5
```

```python
# 6 - Which kind of movies (rating) have a mean duration of more than two hours?

query6 = '''SELECT rating, AVG(length) FROM sakila.film
GROUP by rating
HAVING AVG(length)>120;'''
# pd.set_option('display.max_rows', None)
data6 = pd.read_sql_query(query6, engine)
data6
```

```python
# 7 - Rank films by length (filter out the rows that have nulls or 0s in length column).
# In your output, only select the columns title, length, and the rank


query7 = '''SELECT title, length,
DENSE_RANK () OVER (ORDER BY length DESC) length_rank
FROM sakila.film
WHERE length>0;'''
data7 = pd.read_sql_query(query7, engine)
data7
```
